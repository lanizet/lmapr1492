```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
with MPRester("1d0vsHZ2FSTu7WRT") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-12317")
    # Telecharge le fichier CIF du materiau
    w = CifWriter(structure)
    w.write_file('SrCN2_mp-12317_symmetrized.cif')

#Style ecriture
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))

#Pour les operations de symetrie
from pymatgen.core.operations import SymmOp
#Pour l'illustration des operations de symetrie 
from IPython.display import Image

#Premiere operation de Symetrie : Plan miroir sur le strontium
printmd('**1. Operation de symétrie : Plan miroir défini par (x, x-y, z)**')
OperationSym1 = SymmOp.from_xyz_string('x,x-y,z')
print('La matrice associée à cette opération est la suivante : ')
print(OperationSym1.rotation_matrix)
AtomeSr = OperationSym1.operate((0.33, 0.67, 0.17))
print(' ')
print('Lorsque nous appliquons cette operation de symetrie au Strontium en position (0.33, 0.67, 0.17),')
print('ses coordonnees deviennent : ')
print(AtomeSr)
print(' ')
print('Cette opération de symétrie appliquée au Strontium en position (0.33, 0.67, 0.17)')
print('est illustrée de manière suivante : ')
Image(filename='PlanMiroirSr.png')
    
```

```python
#Deuxième operation de Symetrie appliquee sur le carbone
printmd('**2. Operation de symétrie : Roto-inversion définie par (y, -x+y, -z)**')
OperationSym2 = SymmOp.from_xyz_string('y,-x+y,-z')
print('La matrice associée à cette opération est la suivante : ')
print(OperationSym2.rotation_matrix)
AtomeC = OperationSym1.operate((0.67, 0.33, 0.33))
print(' ')
print('Lorsque nous appliquons cette opération de symétrie au Carbone en position (0.67, 0.33, 0.33),')
print('ses coordonnées deviennent : ')
print(AtomeC)
print(' ')
print('Cette opération de symétrie appliquée au Carbone en position (0.67, 0.33, 0.33)')
print(' est illustrée de manière suivante : ')
Image(filename='RotoInversionC.png')      
```

```python
#Troisieme operation de Symetrie appliquee sur l'azote 
printmd('**3. Operation de symétrie : Translation définie par (x-1/3, y+1/3, z+1/3)**')
OperationSym3 = SymmOp.from_xyz_string('x-1/3,y+1/3,z+1/3')
print('La matrice associée à cette opération est la suivante : ')
print(OperationSym3.rotation_matrix)
print(' ')
print('Lorsque nous appliquons cette opération de symétrie à l azote en position (0.00, 1.00, 0.08),')
print('ses coordonnées deviennent : ')
AtomeN2 = OperationSym3.operate((0.00, 1.00, 0.08))
print(AtomeN2)
print(' ')
print('Cette opération de symétrie appliquée à l azote en position (0.00, 1.00, 0.08)')
print('est illustrée de manière suivante :' )
Image(filename='TranslationN2.png')

```
