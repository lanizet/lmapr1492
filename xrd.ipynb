```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
with MPRester("1d0vsHZ2FSTu7WRT") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-12317")

#Style ecriture
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))    
    
#XRD pattern    
from pymatgen.analysis.diffraction.xrd import XRDCalculator

#Definit la diffraction du rayonnement X incident Cu Ka de longuer d'onde 1,54060 [Amstrong].
xrd = XRDCalculator(1.54060)

#Calcule le XRD pattern de la structure du crystal SrCN2.
pattern = xrd.get_pattern(structure)

#Affiche l'image du diffractogramme avec les indices hkl correspondants à chaque pic.
figure = xrd.get_plot(structure)

#Premier pic du diffractogramme
printmd('**Premier pic du diffractogramme :**')
printmd('*- 2\u03B8 [°]:*')
print(pattern.x[0])
printmd('*- Intensité :*')
print(pattern.y[1])
printmd('*- Indices hkl associés à ce pic :*') 
print(pattern.hkls[0])

#Premier pic du diffractogramme
printmd('**Deuxième pic du diffractogramme :**')
printmd('*- 2\u03B8 [°]:*')
print(pattern.x[1])
printmd('*- Intensité :*')
print(pattern.y[1])
printmd('*- Indices hkl associés à ce pic :*') 
print(pattern.hkls[1])

#Premier pic du diffractogramme
printmd('**Troisième pic du diffractogramme :**')
printmd('*- 2\u03B8 [°]:*')
print(pattern.x[2])
printmd('*- Intensité :*')
print(pattern.y[2])
printmd('*- Indices hkl associés à ce pic :*') 
print(pattern.hkls[2])
```
